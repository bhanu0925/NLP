## Zero Shot classification

```python
zero_shot_clasifier = pipeline("zero-shot-classification")
```

```python
zero_shot_clasifier("this course is about artificial intelligence.",
                    candidate_labels = ["politics","bussiness","education"])
```

## Text generation

```python
generator = pipeline("text-generation")
```

```python
generator("This course is about artificial inteliggence")[0]['generated_text']
```

```python
generator = pipeline("text-generation",model='distilgpt2')
```

```python
generator('This course is about Aartificial intelligence',
          max_length=40,
          num_return_squences=3)
```

```python
generator('The movie was horrible',
          max_length=40,
          num_return_squences=4)
```

```python
generator("<p> this is html </p>",
                    max_length=40,
          num_return_squences=4)
```

```python
unmasking = pipeline("fill-mask")
```

##UNMask

# NER

```python
NER = pipeline("ner")
```

```python
NER(" I am Bhanu and i work at TCS and i live in Bangalore  ",grouped_entities = True)
```

#QA

```python
QA = pipeline("question-answering")
```

```python
QA(question = "where do i work",
   context ="I am Bhanu and i work at TCS and i live in Bangalore ")
```

## Summerization

```python
summerization = pipeline("summarization")
```
